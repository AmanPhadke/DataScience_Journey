```python
model = YOLO('yolov8n.pt')

path = '/content/Shelf_1.jpg'
results = model.predict(path, save = True)
```

```python
items = []
boxes = results[0].boxes

for class_id in boxes.cls:
  key = int(class_id.item())
  item = model.names[key]
  if item not in items:
    items.append(item)

print(items)
```
