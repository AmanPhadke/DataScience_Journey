```python
import os, shutil
```

```python
path = r'' #Add your directory path which you want to sort
```

```python
file_name = os.listdir(path)
```

```python
file_extentions = []
for ext in file_name:
    if 'iles' not in ext[-4:]:
        file_extentions.append(ext[-4:])
```

```python
extentions = []
for i in file_extentions:
    if i not in extentions:
        extentions.append(i)
```

```python
folders = []

for e in extentions:
    folder = e.replace('.', '') + ' files'
    folders.append(folder)
```

```python
for i in range(0, len(folders)):
    if not os.path.exists(path + folders[i]):
        os.makedirs(path + folders[i])
```

```python
print(folders)
print(extentions)
print(file_name)
```

```python
for file in file_name:
    for ext in extentions:
        for folder in folders:
            if ext in file and ext[-3:] in folder and not os.path.exists(path + folder + '/' + file):
                shutil.move(path + file, path + folder +'/'+ file)
```
